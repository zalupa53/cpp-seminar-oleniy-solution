**Правило №2: Работайте с данными по частям (батчи)**

Не пытайтесь загрузить весь датасет в память GPU целиком, если он большой. Это самый быстрый способ получить ошибку `CUDA out of memory`. Решение здесь очевидное — батчинг (batching). Данные подаются в модель небольшими порциями (батчами). Это стандартная практика в Deep Learning, и все современные фреймворки работают именно так через `DataLoader`.

`DataLoader` - это, по сути, умный и эффективный поставщик данных для вашей модели. Он берет ваш огромный `Dataset`, нарезает его на небольшие батчи и подает их по одному. Кроме того, он умеет перемешивать данные перед каждой эпохой, чтобы модель не заучивала порядок, и использовать несколько процессорных ядер для загрузки следующего батча, пока GPU занят текущим.

```python
# Пример DataLoader для PyTorch
from torch.utils.data import DataLoader
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Если доступна CUDA, будет смысл в pin_memory и non_blocking копировании
use_cuda = torch.cuda.is_available()

loader = DataLoader(
    dataset,
    batch_size=64, # размер батча
    shuffle=True, # для обучения — True; для валидации/инференса — False
    num_workers=2, # параллельная подготовка данных на CPU
    pin_memory=use_cuda, # фиксированная память: быстрее копировать батчи на GPU
    persistent_workers=True if 2 > 0 else False,  # сохраняет процессы-загрузчики между эпохами, убирая задержки на их создание в начале каждой эпохи
)

# В цикле лучше копировать батчи на устройство с non_blocking при включенном pin_memory:
# for x, y in loader:
#     x = x.to(device, non_blocking=use_cuda)
#     y = y.to(device, non_blocking=use_cuda)
```

> `pin_memory` ускоряет копирование на GPU, особенно вместе с `tensor.to(device, non_blocking=True)`. Эффекта не будет, если вы не переносите батчи на `CUDA`.
>
> Подбирайте `batch_size` под VRAM/ОЗУ. Псс, если вы уже большой профи и хочется юзать батч побольше - почитайте про *градиентную аккумуляцию*.

**Правило №4: Сохраняйтесь!**

Даже при соблюдении всех правил что-то может пойти не так: случайная ошибка в коде, окончание лимита времени сессии. Чтобы не потерять часы прогресса, внедрите в свой код чекпоинты (checkpoints).

* **Регрессия** - предсказываем именно число (непрерывное значение).

**Метрики в наших задачах**

Для "Мыслим категориями" это **HitRate@3 (H@3)**.

Смысл: попала ли хотя бы одна правильная категория в ваш топ-3? Считаем долю таких объектов:
$$ H@3 = \frac{\text{объекты, где хотя бы одна верная метка в топ-3}}{\text{все объекты}} $$

*Интуитивно*. Если часто заходим хотя бы одной меткой в тройку - метрика высокая.

```python
def hitrate_at_3(true_labels, pred_top3):
    hits = 0
    for true_labels, top3 in zip(true_labels, pred_top3):
        # есть ли пересечение между множеством истинных меток и предсказанных
        hits += int(any(lbl in top3 for lbl in true_labels))
    return hits / len(true_labels)
```

Для нашей второй задачи "АвтоПрайс" - **medianAPE** (медиана относительной ошибки)

Сначала для каждого автомобиля считаем относительную ошибку **APE** (Absolute Percentage Error):

$$ APE_i = \frac{| \hat{y}_i - y_i |}{y_i},$$

где $y_i$ - настоящая цена, а $\hat{y}_i$ - предсказанная. Затем просто берём медиану по всем объектам:

$$\text{medianAPE} = \operatorname{median}_i(APE_i)$$

> Вы резонно спросите: "Почему медиана, а не среднее?" Медиана гораздо более устойчива к выбросам. Если модель на нескольких очень дорогих или необычных машинах ошиблась очень сильно, на медиане это скажется слабо, а среднее значение испортится.

```python
import numpy as np

def median_ape(y_true, y_pred):
    # numpy-массивы будут удобнее
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    ape = np.abs(y_pred - y_true) / y_true
    return float(np.median(ape))
```

Разумеется, это не единственные существующие метрики: там ценый зоопарк. Обязательно ощнакомьтесь на досуге. А пока надо обсудить значимость `валидации`.

**Как делим данные: train / validation**

Давайте здесь же рассмотрим еще один очень важный момент: выделение валидации. Мы же хотим проверять, насколько наша модель обобщает, а не просто запоминает данные. Поэтому:

`train` - здесь учим модель;

`validation` - тут подбираем настройки/модель, контролируем *переобучение* (подгонку под данные и потерю обобщающей способности) и чекаем метрики.

Практика для наших задач:

- "Мыслим категориями". Можно сделать простой *случайный сплит* (например, 80/20). Если хотите аккуратнее - стремитесь, чтобы распределение меток в train и val было похоже. Для мульти-меток это делается специальной *итерированной стратификацией*, но пока можно обойтись и простым сплитом.
- "АвтоПрайс". В задании тест - это декабрь 2024, а трейны - до ноября. Если делаете локальную валидацию, имитируйте такую же схему: например, `train` = январь–октябрь, а `val` = ноябрь. Это честнее по времени и помогает избежать утечки будущего в прошлое.

**Три важных правила**

1. Никакой "утечки". Все преобразования, которые учатся на данных (например, нормализация/векторизация текста), `fit` только на `train`, а к `val` применяем `transform`.
2. Фиксируйте случайность (`seed`), чтобы результаты воспроизводились.
3. Ранний стоп и простые бенчмарки: сравнивайте с бейзлайном и не перетюньте модель под валидацию.

### Модели: классические и нейронные

Итак, мы определились с типом задачи (классификация или регрессия) и подготовили данные. Теперь нужно выбрать инструмент для решения - модель. Представьте, что у вас есть ящик с инструментами: для одних задач лучше подойдет молоток, для других - отвертка. С моделями так же. Их выбор в первую очередь зависит от типа ваших данных.

**Классические модели**

Если ваши данные выглядят как таблица (строки - объекты, столбцы - признаки), то ваш выбор - классические модели машинного обучения.

Это проверенные временем и очень мощные алгоритмы, которые отлично находят закономерности в структурированных данных. Вот главные из них:

- **Линейная и логистическая регрессии** (`LinearRegression`, `LogisticRegression`). Простые, быстрые и понятные модели. Идеальны как бейзлайн - первое решение, с которым вы будете сравнивать все более сложные подходы. Требуют грамотного подбора гиперпараметров.

- **Случайный лес** (`RandomForest`). Более мощная модель, которая строит много деревьев решений и усредняет их предсказания. Хорошо работает и "из коробки".

- **Градиентный бустинг** (`Gradient Boosting`). Настоящий чемпион соревнований на табличных данных. Такие библиотеки, как `CatBoost`, `LightGBM` и `XGBoost`, реализуют эту идею. Бустинг последовательно строит ансамбль моделей, где каждая следующая исправляет ошибки предыдущих. Если у вас табличные данные, бустинг - первое, что стоит попробовать для получения высокого результата.

**Нейронные сети**

А что делать, если данные - это не таблица, а что-то более сложное? Как например в наших отборочных задачах: тексты или картинки? Классические модели не могут работать с ними напрямую.

Вот здесь нам и пригодится **Deep Learning**. Суперспособность нейронки - превращать сложные объекты в набор чисел (вектор), сохраняя при этом их смысл. Этот вектор называют `эмбеддингом` или представлением (representation).

Нейросеть выступает в роли универсального переводчика:

- Для текстов: модели-трансформеры (`BERT`-подобные, `T5` и др.) читают текст и выдают вектор, который отражает его семантику.

- Для картинок: сверточные сети (`ResNet`, `EfficientNet`, `ConvNeXt`) и `ViT`. Дают эмбеддинги картинки.

Получив эти числовые представления для всех наших объектов, мы фактически снова получаем - таблицу! А с таблицами мы уже знаем, что делать. Но и тут у нас возникает две стратегии.

**Два практических маршрута**

Итак, нейросеть выдала нам качественные эмбеддинги для наших текстов/картинок. Дальше есть две основные стратегии:

**Вариант 1. Эмбеддинги → классическая модель**

Это двухэтапный подход:

**Этап 1.** Берем мощную предобученную нейросеть и пропускаем через нее все наши тексты/картинки. На выходе для каждого объекта получаем эмбеддинг. Сохраняем эти векторы.

**Этап 2.** Обучаем на этих векторах наш любимый градиентный бустинг (например, CatBoost), который и будет делать финальное предсказание.

Плюсы:

- Быстро стартует, мало кода, экономно к GPU (часто хватит T4/CPU).

- Данные нужны умеренно, качество часто уже приличное.

- Удобно склеивать признаки: эмбеддинг + табличные поля → бум, отправляем сразу в бустинг.

Минусы:

- Качество ограничено - "как умеет" предобученный энкодер.

- Надо тщательно подбирать размер/тип эмбеддинга.



**Вариант 2. Нейросеть End-to-End**

В этом подходе мы не разделяем процесс на два этапа. Мы просто добавляем к нашей нейросети еще один, последний слой - "голову". Эта голова, по сути, и есть простая логистическая/линейная регрессия.

Вся эта конструкция (огромный трансформер + маленькая "голова") обучается целиком, от начала и до конца (end-to-end).

Плюсы:

- Потенциально выше потолок качества: сеть подстраивается под вашу задачу.

- Можно тонко учить "что важно" конкретно для вашего датасета.

Минусы:

- Нужнее GPU и аккуратная настройка (LR, аугментации, ранний стоп).

- Риск переобучения на маленьких данных выше.

- Линейный слой - это не бустинг. Он может упустить какие-то сложные зависимости, которые бустинг бы нашел.

- Дольше по времени, больше "подводных камней".

| Ситуация                                           | Что попробовать                                           |
| -------------------------------------------------- | --------------------------------------------------------- |
| Мало времени/ресурсов                              | Эмбеддинг → бустинг                                       |
| Много табличных признаков                          | Бустинг как основной инструмент                       |
| Нужно быстро получить стабильный baseline          | TF-IDF + логрег (текст) / бустинг (таблица)               |
| Данных хватает, нужен максимум качества            | End-to-end дообучение + аугментации + ранний стоп         |
| Есть и табличка, и сырые данные (текст/картинка) | Эмбеддинг (текст/картинка) + табличные фичи → бустинг |

**Ансамбли и AutoML**

Когда вы уже попробовали разные модели, есть два верных способа улучшить результат еще больше.

**Один в поле не воин**

**Ансамбль** - это "комитет" из нескольких моделей, которые принимают решение совместно. Идея простая: объединяем предсказания разных моделей, чтобы сгладить их ошибки.

- `Bagging` - учим несколько одинаковых моделей на разных подвыборках (пример: RandomForest).

- `Boosting` - модели учатся последовательно, каждая "докручивает" ошибки предыдущей. Это как раз наши топ-модельки: CatBoost/LightGBM/XGBoost. Часто лучший выбор на табличках.

- `Blending` - простое смешивание предсказаний (среднее/взвешенное среднее).

- `Stacking` - учим мета-модель на предсказаниях базовых моделей. Требуется аккуратная валидация: OOF-предикты.

Когда помогает: модели разнообразны (другие фичи, другая архитектура, другие гиперпараметры/seed). Если все модели одинаковые - толку мало.

> Вы заметили, что в одно предложение эти подходы не упаковать? Обязательно займитесь погружением в эту тему - там таится *золото*.

**Автоматизация поиска лучшей модели**

**AutoML** (Автоматическое машинное обучение) - это инструменты, которые берут на себя всю рутинную работу по подбору моделей и их настроек.

Представьте себе робота, которому вы даете на вход данные и говорите: "Найди мне лучшую модель для этой задачи". Робот сам попробует и линейную регрессию, и случайный лес, и бустинг, переберет их самые важные параметры и в конце выдаст вам готовое решение-чемпион.

Популярные библиотеки: `AutoGluon`, `LightAutoML`, `PyCaret`.

Когда использовать?

- Для быстрого старта. AutoML может за несколько часов создать очень сильный бейзлайн, который будет сложно превзойти вручную.

- Для поиска идей. Иногда AutoML находит такие комбинации признаков или моделей, до которых вы бы сами не додумались.

> Важно понимать, что AutoML - это не "волшебная кнопка", а очень мощный помощник. Он не отменяет необходимости понимать, как работают модели, но может значительно ускорить ваш путь к качественному решению.

Итог: начните решение с одной сильной модели (бустинг на табличке / эмбеддинг → бустинг), затем добавьте 1–2 разнообразные модели и простой blending. Если нужно - подключите AutoML с ограничением времени. Это даст рост качества при контролируемых ресурсах.

Вы просто приходите, выбираете лучшую модель (например, `BERT`), ставите ее в свой "автомобиль" (`PyTorch`-модель) и получаете фантастический результат.

```python
# Пример кода "текст → эмбеддинги"

from transformers import AutoTokenizer, AutoModel
import torch, numpy as np

model_id = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2" # Пример многоязычной модели
tok = AutoTokenizer.from_pretrained(model_id)
enc = AutoModel.from_pretrained(model_id).eval().to(device)

def text_embed(texts, batch=64):
    out = []
    for i in range(0, len(texts), batch):
        t = tok(texts[i:i+batch], padding=True, truncation=True, return_tensors="pt").to(device)
        with torch.no_grad():
            h = enc(**t).last_hidden_state # (B, L, H)
            emb = h.mean(dim=1) # Простое mean-pooling
        out.append(emb.cpu().numpy())
    return np.vstack(out)

X_text = text_embed(apps["text"].tolist())
```

### Работа с текстом

Ну вот мы и подошли к самому интересному: как *превратить слова в числа* под нашу задачу. У нас есть названия и описания приложений, а у машины - умение работать с числами. Задача этого этапа - построить мост между этими двумя мирами.

Процесс превращения текста в осмысленный числовой вектор называется `векторизацией`. Давайте посмотрим на основные способы это сделать, от простых к самым мощным.

**1. "Мешок слов" (Bag-of-Words) и TF-IDF**

Это классический и очень интуитивный подход.

Представьте, что мы берем описание приложения и высыпаем все слова из него в один большой мешок, перемешиваем и считаем, сколько раз каждое слово встретилось. Порядок слов мы полностью игнорируем. На основе этих подсчетов для каждого текста создается длинный вектор, где каждая позиция отвечает за конкретное слово из общего словаря всех приложений.

Часто вместо простых подсчетов используют `TF-IDF` - метрику, которая показывает, насколько слово важно для конкретного текста в контексте всей коллекции текстов.

- Плюсы. Просто, быстро, не требует GPU. Отлично работает как первый бейзлайн и хорошо справляется с определением общей тематики текста.

- Минусы. Полностью теряется контекст и порядок слов. Для модели фразы "ребенок играет в телефон" и "телефон играет в ребенок" - одно и то же. Не понимает синонимов ("игра" и "game" - разные слова).

**2. Контекстные эмбеддинги (BERT и Трансформеры)**

Это современный и самый мощный подход, который мы и будем использовать.

Модели-трансформеры, такие как BERT, читают текст целиком, обращая внимание на контекст каждого слова. Они не просто переводят слова в векторы, а создают динамические представления, зависящие от окружения. На выходе модель выдает готовый вектор (эмбеддинг) для всего текста, который "впитал" в себя его смысл.

- Плюсы. State-of-the-Art (SOTA) результаты почти во всех задачах NLP. Отлично улавливает нюансы, сарказм, сложные зависимости в тексте. Понимает, что синонимы близки по смыслу.

- Минусы. Требует значительных вычислительных ресурсов (GPU обязателен). Работает медленнее, чем классические подходы.

**Как выбрать правильный "BERT"?**

Мы взяли название семейства моделей в ковычки, поскольку на деле нам нужен не Берт - а именно `энкодер-модель`. Это модет быть и e5, и энкодер часть от T5. Это непринципиально. Главное тут - сделать правильный выбор модели.

- **Язык.** Первое и главное правило. Для нашей задачи с русскими описаниями нужно использовать модель, предобученную на огромном корпусе русских текстов. Использование английского BERTа даст ужасные результаты.

- **Размер и специфика.** Существуют модели разного размера (base, large) и дообученные под конкретные задачи.

- **Используйте лидерборды.** Лучший способ выбрать - зайти на Hugging Face ([MTEB](https://huggingface.co/spaces/mteb/leaderboard) - Embedding Leaderboard, не забудьте переключить на русский сегмент) или посмотреть на лидерборды для русского языка (например, [Russian SuperGLUE](https://russiansuperglue.com/leaderboard/2)). Там модели соревнуются на разных задачах (включая классификацию), и вы можете выбрать ту, что показывает лучшие результаты.

> Обратили уже внимание, что эти два лидерборда отличаются? А знаете почему? Один из них - для моделей, которые делают `семантические эмбеддинги`. В прошлом году мы уже знакомились с такими.

Давайте разберемся, в чем тут суть.

- **Russian SuperGLUE** - это комплексный экзамен на "общее понимание языка". Модели на этом лидерборде соревнуются в решении разнообразных задач: ответить на вопрос, определить, следует ли одно предложение из другого, и, в том числе, классифицировать тексты. Модели-чемпионы здесь - это отличные заготовки для дальнейшего дообучения (`fine-tuning`) под вашу конкретную задачу. Это идеально подходит для нашего "Вариает 2: Нейросеть 'под ключ'".

- **MTEB** (Massive Text Embedding Benchmark), в свою очередь, проверяет совсем другое. Он оценивает, насколько качественными получаются выходные векторы (эмбеддинги) "сами по себе". Главный критерий здесь - сохраняют ли эти векторы смысл исходного текста. Этот лидерборд отвечает на вопрос: "Если я просто прогоню два текста через модель и сравню их векторы, смогу ли я понять, похожи ли тексты по смыслу?"

Именно такие модели, которые нацелены на создание качественных векторов для сравнения, и делают те самые `семантические эмбеддинги`.

**Когда именно нам нужны семантические эмбеддеры?**

Семантические эмбеддинги - это ваш инструмент № 1, когда задача требует понимания и сравнения смысла на уровне целых предложений или абзацев.

Вам стоит смотреть на **MTEB** и выбирать модели из его топа, если ваша задача - это:

- **Семантический поиск.** Найти документы, наиболее близкие по смыслу к вашему запросу, даже если в них нет тех же ключевых слов. (Например, по запросу "недорогие телефоны с хорошей камерой" найти текст "обзор бюджетных камерофонов").

- **Кластеризация текстов.** Сгруппировать похожие по смыслу новости, отзывы или документы.

- **Поиск дубликатов или плагиата**. Найти два текста, которые говорят об одном и том же разными словами.

- **Системы рекомендаций.** Порекомендовать пользователю статьи или товары, похожие на те, что ему уже понравились.

*Как это относится к нашей задаче?*

Наша задача - классификация, а не поиск. Поэтому, на первый взгляд, нам важнее лидерборд `Russian SuperGLUE`. Модель из его топа будет отличной основой для дообучения (fine-tuning) под нашу задачу с категориями.

НО! Если мы решим пойти по "Вариант 1" (Нейросеть + Бустинг), то качество эмбеддингов становится для нас критически важным. Ведь мы сначала превращаем все тексты в векторы, а потом на этих векторах обучаем CatBoost. В этом сценарии модель-чемпион из MTEB может дать нам более качественные "признаки" для бустинга, чем модель из топа SuperGLUE.

> **Вывод.** Оба лидерборда полезны. Russian SuperGLUE помогает выбрать лучшую "базу" для дообучения, а MTEB - лучший "генератор признаков-эмбеддингов" для последующей работы с классическими моделями.
>
> Но будьте внимательны. Выбор модели еще и зависит от размера данных: BERT отработает на малых текстах лучше, чем SBERT. Поэтому обязательно проверяйте валидность разных типов энкодеров.

```python
expected_cols = {"app_name", "full_description", "shortDescription", "labels_str"}

def parse_labels(s):

    """Преобразует строку вида 'cat1|cat2' в список ['cat1', 'cat2']"""

    if pd.isna(s):
       return []
    return [t.strip() for t in str(s).split("|") if t.strip()]

df_train["labels_list"] = df_train["labels_str"].apply(parse_labels)
df_train["n_labels"] = df_train["labels_list"].apply(len)
all_labels = [label for labels in df_train["labels_list"] for label in labels]
label_counts = Counter(all_labels) # сколько раз встречается каждая уникальная метка

labels_df = (pd.DataFrame({"Категория": list(label_counts.keys()), "Количество": list(label_counts.values())})
             .sort_values("Количество", ascending=False)
             .reset_index(drop=True))
labels_df["Доля_%"] = (labels_df["Количество"] / len(df_train) * 100).round(2)
card_df = (df_train["n_labels"].value_counts().sort_index()
           .rename_axis("Категорий на приложение").reset_index(name="Количество приложений"))
card_df["Доля_%"] = (card_df["Количество приложений"] / len(df_train) * 100).round(2)

print(f"Всего приложений в датасете: {len(df_train)}")
print(f"Всего уникальных категорий: {labels_df.shape[0]}")
print(f"Среднее число категорий на приложение: {df_train['n_labels'].mean():.2f}")
```

```python
# Еще давайте глянем на совместную встречаемость лейблов (Co-occurrence Matrix)

TOPN = 20 # Возьмем топ-20 категорий для наглядности
topN_labels = labels_df["Категория"].head(TOPN).tolist()
label_to_idx = {label: i for i, label in enumerate(topN_labels)}

co_occurrence_matrix = np.zeros((TOPN, TOPN), dtype=int)

for labels in df_train["labels_list"]:
    # Фильтруем метки, оставляя только те, что входят в наш топ
    top_labels_in_app = [l for l in labels if l in label_to_idx]
    # Проходим по всем парам меток в одном приложении
    for i in range(len(top_labels_in_app)):
        for j in range(i, len(top_labels_in_app)):
            idx1 = label_to_idx[top_labels_in_app[i]]
            idx2 = label_to_idx[top_labels_in_app[j]]
            co_occurrence_matrix[idx1, idx2] += 1
            # Матрица симметрична, поэтому заполняем и вторую половину
            if idx1 != idx2:
                co_occurrence_matrix[idx2, idx1] += 1

plt.figure(figsize=(12, 8))
sns.heatmap(co_occurrence_matrix, annot=True, fmt="d", cmap="YlGnBu",
            xticklabels=topN_labels, yticklabels=topN_labels)
plt.title(f"Как часто топ-{TOPN} категорий встречаются вместе?", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Подготовка данных

После небольшого EDA (разведочного анализа данных), давайте перейдем к подготовке данных: текста и лейблов.

Так как это базовое решение - не будем ничего "мудрить" и просто склеим тексты.

```python
def expand_multi_label_examples(df):
    """
    Создает дубликаты для примеров с несколькими метками.
    Каждая строка с N метками превращается в N строк с одной меткой.
    """
    expanded_rows = []

    for idx, row in df.iterrows():
        labels_list = row["labels_list"]

        if len(labels_list) == 0:
            # Если нет меток, добавляем строку с меткой "unknown"
            new_row = row.copy()
            new_row["primary_label"] = "unknown"
            expanded_rows.append(new_row)
        elif len(labels_list) == 1:
            # Если одна метка, оставляем как есть
            new_row = row.copy()
            new_row["primary_label"] = labels_list[0]
            expanded_rows.append(new_row)
        else:
            # Если несколько меток, создаем дубликаты для каждой метки
            for label in labels_list:
                new_row = row.copy()
                new_row["primary_label"] = label
                expanded_rows.append(new_row)

    expanded_df = pd.DataFrame(expanded_rows)
    return expanded_df

# Создаем расширенный датасет с дубликатами
df_train_expanded = expand_multi_label_examples(df_train)

print(f"Исходный размер датасета: {len(df_train)}")
print(f"Расширенный размер датасета: {len(df_train_expanded)}")

# Статистика по дубликатам
original_label_counts = df_train["n_labels"].value_counts().sort_index()
expanded_label_counts = df_train_expanded["primary_label"].value_counts()

print("\nСтатистика по количеству меток в исходных данных:")
for n_labels, count in original_label_counts.items():
    print(f"  {n_labels} меток: {count} примеров")

print(f"\nВсего уникальных меток в расширенном датасете: {len(expanded_label_counts)}")

# Преобразуем в числовые метки для классификации
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_primary = label_encoder.fit_transform(df_train_expanded["primary_label"])
NUM_CLASSES = len(label_encoder.classes_)

print(f"\nЧисло классов: {NUM_CLASSES}")
print(f"Пример primary labels: {df_train_expanded['primary_label'].head(8).tolist()}")
print(f"Пример encoded labels: {y_primary[:8].tolist()}")

# Дополнительная информация о распределении
print(f"\nТоп-10 самых частых меток:")
top_labels = df_train_expanded["primary_label"].value_counts().head(10)
for label, count in top_labels.items():
    print(f"  {label}: {count} примеров")
```

Что делает [MultiLabelBinarizer](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MultiLabelBinarizer.html)?

Это удобный "кодировщик" для многометочных задач: у каждого приложения есть список категорий (например, `["tools", "puzzle"]`), а модели нужен числовой вектор фиксированной длины.

`MultiLabelBinarizer()` строит словарь всех встречающихся меток и переводит каждый список в мульти-хот-вектор: 1 - метка присутствует, 0 - нет. Порядок столбцов (классов) фиксируется внутри объекта `mlb` (мы сохраняем его в `labels.json`), поэтому на инференсе важно использовать тот же порядок, чтобы правильно вернуть названия категорий из индексов. Обратное преобразование делается через `mlb.inverse_transform(...)`, а список классов доступен в `mlb.classes_`.

Итак, мы подготовили наши данные: склеили тексты и разобрались с метками. Но нейросеть по-прежнему не может "прочитать" текст "*Run and Jump - Бесконечный раннер*". Модели вроде `BERT` понимают не слова, а последовательности чисел.

`Токенайзер` - это наш переводчик с человеческого языка на язык, понятный для нейросети. Он делает три важные вещи:

1. Разбивает текст на "кирпичики" (токены). Он не просто делит текст по пробелам. Умные токенизаторы разбивают слова на осмысленные части (`subwords`). Например, слово "нейросеть" может превратиться в два токена: ["нейро", "##сеть"]. Это помогает модели понимать структуру слов и работать даже с теми словами, которых она раньше не видела.

2. Переводит токены в числа. У каждого токенизатора есть свой "словарь" (`vocabulary`), где каждому токену присвоен уникальный номер (ID). Например, ["нейро", "##сеть"] может превратиться в [15345, 18656].

3. Добавляет служебные символы. Модели `BERT` для работы нужны специальные токены, например [CLS] в начале текста (чтобы "собирать" общую информацию о предложении) и [SEP] в конце.

Еще и дико удобно, что токенайзер можно просто подтянуть с HF вместе с моделью. Каждая модель требует строго свой токерайзер.

```python
def create_single_label_model():
    return AutoModelForSequenceClassification.from_pretrained(
        MODEL_ID,
        num_labels=NUM_CLASSES,
        problem_type="single_label_classification",  # Важно: single_label!
    )

# Обновляем Dataset для single-label
class SingleLabelDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels=None):
        self.enc = encodings
        self.labels = labels

    def __len__(self):
        return len(self.enc["input_ids"])  # Важно: возвращаем длину

    def __getitem__(self, i):
        item = {k: torch.tensor(v[i]) for k, v in self.enc.items()}
        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[i]).long()
        return item
```

```python
def hitrate_at_k_single(logits, y_true, k=3):
    """
    Hits@K для single-label classification
    Считает, попадает ли истинная метка в топ-K предсказаний
    """
    # Получаем индексы топ-K предсказаний (по убыванию уверенности)
    topk_indices = np.argsort(-logits, axis=1)[:, :k]

    # Проверяем, есть ли истинная метка в топ-K
    hits = 0
    for i in range(len(y_true)):
        if y_true[i] in topk_indices[i]:
            hits += 1

    return hits / len(y_true)

def compute_metrics_single(eval_pred):
    logits, labels = eval_pred
    return {"h3": hitrate_at_k_single(logits, labels, k=3)}
```

### Модель: BERT + "голова" для multi-label

Теперь, когда все данные подготовлены, мы наконец-то можем собрать и обучить нашу модель. Этот процесс состоит из четырех основных этапов:

1. Сборка модели. Мы берем предобученную "тушку" `BERT` и прикрепляем к ней "голову" для нашей конкретной задачи.

2. Создание "линейки". Мы пишем функцию для подсчета нашей метрики `Hitrate@3`, чтобы понимать, насколько хорошо модель справляется.

3. Настройка "режима тренировки". Мы задаем все параметры обучения - сколько эпох учиться, с какой скоростью, какими порциями и т.д.

4. Запуск "тренера". Мы передаем все компоненты специальному объекту `Trainer`, который берет всю сложную работу на себя.

**Собираем нашу модель**

Используем `AutoModelForSequenceClassification` - это специальный класс из Hugging Face, который делает именно то, что нам нужно: берет базовую модель-трансформер (наш `MODEL_ID`) и автоматически добавляет сверху один слой-классификатор (ту самую "голову").

```python
def create_model():
  return  AutoModelForSequenceClassification.from_pretrained(
      MODEL_ID,
      num_labels=NUM_LABELS,
      problem_type="multi_label_classification",  # включает BCEWithLogitsLoss + sigmoid - вот тут
  )
```

**Готовим "линейку" для оценки**

`Trainer`'у нужно объяснить, как считать качество модели во время валидации. Для этого мы пишем две функции:

- `hitrate_at_k`: основная логика, которая считает `Hitrate@3`. Она берет сырые выходы модели (логиты), находит 3 самых вероятных предсказания и проверяет, есть ли среди них хотя бы одно правильное.

- `compute_metrics`: простая "обертка", которая получает от `Trainer`'а предсказания и правильные метки и передает их в нашу функцию `hitrate_at_k`.

```python
def hitrate_at_k(logits, y_true, k=3):
    topk = np.argpartition(-logits, kth=k-1, axis=1)[:, :k]

    hits = 0
    total_samples = len(y_true)

    for i in range(total_samples):
        true_labels = np.where(y_true[i] == 1)[0]
        predicted_topk = topk[i]

        # Отладочная информация для первых нескольких примеров
        if i < 3:
            print(f"Sample {i}:")
            print(f"  True labels: {true_labels}")
            print(f"  Top-{k} predictions: {predicted_topk}")
            print(f"  Intersection: {set(true_labels) & set(predicted_topk)}")

        if len(set(true_labels) & set(predicted_topk)) > 0:
            hits += 1

    hit_rate = hits / total_samples

    print(f"Debug: hits={hits}, total={total_samples}, hit_rate={hit_rate:.4f}")
    return hit_rate
```

### Инференс и сабмит

"Инференс" (inference) - это просто стильное словечко, которое означает "применение обученной модели" к новым данным, которых она раньше не видела. Наша модель готова, и теперь мы применим ее к тестовой выборке (`test.tsv`), чтобы получить предсказания и сформировать итоговый файл.

```python
# загружаем тест
df_test = pd.read_csv("test.tsv", sep="\t")
df_test["text"] = df_test.apply(build_text, axis=1)

# токенизируем
te_enc = tokenize_texts(df_test["text"].tolist())
test_ds = Dataset(te_enc, labels=None)

# предсказания (логиты)
pred_logits = trainer.predict(test_ds).predictions

# берём top-3 меток и сортируем их по убыванию уверенности
id2label = np.array(labels)
top3_idx = np.argpartition(-pred_logits, kth=2, axis=1)[:, :3]

# упорядочим эти три по логитам
ordered = np.take_along_axis(
    top3_idx,
    np.argsort(-np.take_along_axis(pred_logits, top3_idx, axis=1), axis=1),
    axis=1
)

# gреобразуем отсортированные индексы в строки с метками, разделенными '|'
pred_top3 = ["|".join(id2label[row]) for row in ordered]

# DataFrame для сабмита
submission = pd.DataFrame({
    "app_name": df_test["app_name"],
    "labels_str": pred_top3
})
submission.to_csv("submission.tsv", sep="\t", index=False)
submission.head()
```
