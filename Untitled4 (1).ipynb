```python
train = pd.read_csv("/content/train (33).csv", sep=";")
train = train.fillna('unknown')
```

```python
target_labels = [
    'Развлечения и юмор', 'Кулинария', 'Торговля и объявления', 'СМИ',
    'Философия и религия', 'Животные', 'Творчество и дизайн', 'Путешествия'
]

X_train = train[train['label'].isin(target_labels)].reset_index(drop=True)
```

```python
def func(x):
  label2idx = {
    'Животные': 0,
    'Кулинария': 1,
    'Путешествия': 2,
    'Развлечения и юмор': 3,
    'СМИ': 4,
    'Творчество и дизайн': 5,
    'Торговля и объявления': 6,
    'Философия и религия': 7
  }
  return label2idx[x]
def func_back(x):
  idx2label = {
      0: 'Животные',
      1: 'Кулинария',
      2: 'Путешествия',
      3: 'Развлечения и юмор',
      4: 'СМИ',
      5: 'Творчество и дизайн',
      6: 'Торговля и объявления',
      7: 'Философия и религия'
  }
  return idx2label[x]
```

```python
X_train['mapped'] = X_train['label'].map(func)
```

```python
class ImageDataset(torch.utils.data.Dataset):
    def __init__(self, dataframe, indices, photo_path, transform, train=True):
        self.dataframe = dataframe
        self.transform = transform
        self.train = train
        self.indices = indices
        self.photo_path = photo_path

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        original_idx = self.indices[idx]
        filename = self.dataframe.loc[original_idx, 'id']
        file_path = os.path.join(self.photo_path, str(filename))

        image = Image.open(file_path).convert("RGB")
        image = self.transform(image)

        if self.train:
            label = torch.tensor(self.dataframe['mapped'].loc[original_idx], dtype=torch.long)
            return image, label
        else:
            return image
```

```python
transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            ),
        ])
```

```python


class Resnet(nn.Module):
    def __init__(self, num_classes=8):
        super(Resnet, self).__init__()

        self.model = models.resnet152(weights=models.ResNet152_Weights.IMAGENET1K_V2)

        self.model.fc = nn.Linear(self.model.fc.in_features, num_classes)

    def forward(self, x):
        return self.model(x)

class EfficientNet(nn.Module):
  def __init__(self, num_classes=8):
    super(EfficientNet, self).__init__()

    self.model = models.efficientnet_v2_m(weights=models.EfficientNet_V2_M_Weights.IMAGENET1K_V1)

    self.model.classifier[1] = nn.Linear(self.model.classifier[1].in_features, num_classes)
  def forward(self, x):
    return self.model(x)
class EfficientNetLarge(nn.Module):
    def __init__(self, num_classes=8):
        super(EfficientNetLarge, self).__init__()

        self.model = models.efficientnet_v2_l(weights=models.EfficientNet_V2_L_Weights.IMAGENET1K_V1)

        in_features = self.model.classifier[1].in_features
        self.model.classifier[1] = nn.Linear(in_features, num_classes)

    def forward(self, x):
        return self.model(x)
```

```python
from sklearn.model_selection import train_test_split

train_idx, val_idx = train_test_split(
    X_train.index.tolist(),
    test_size=0.2,
    stratify=X_train['mapped'],
    random_state=42
)


photo_path = "/content/Data (1)/Train/"
test_photo_path = "/content/Data (1)/Test/"

train_dataset = ImageDataset(X_train, train_idx, photo_path, transform, train=True)
val_dataset = ImageDataset(X_train, val_idx, photo_path, transform, train=True)
train_full_dataset = ImageDataset(X_train, X_train.index.tolist(), photo_path, transform, train=True)
```

```python
test_df = pd.read_csv("test (37).csv", sep=';')
test = test_df.copy()
test_idx = test_df.index.tolist()
test_dataset = ImageDataset(test_df, test_idx, test_photo_path, transform, train=False)
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
train_full_loader = DataLoader(train_full_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)
```

```python
def train_model(model, train_loader, val_loader, optimizer, device, epochs=3):
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        train_preds = []
        train_labels = []

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            train_labels.extend(labels.cpu().numpy())

        train_f1 = f1_score(train_labels, train_preds, average="macro")
        print(f"[Epoch {epoch+1}] Train Loss: {train_loss/len(train_loader):.4f}, F1: {train_f1:.4f}")

        model.eval()
        val_loss = 0.0
        val_preds = []
        val_labels = []

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss += loss.item()
                val_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
                val_labels.extend(labels.cpu().numpy())

        val_f1 = f1_score(val_labels, val_preds, average="macro")
        print(f"[Epoch {epoch+1}] Val Loss: {val_loss/len(val_loader):.4f}, F1: {val_f1:.4f}")
```

```python
def train_model_without_eval(model, train_loader, optimizer, device, epochs=3):
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        train_preds = []
        train_labels = []

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            train_labels.extend(labels.cpu().numpy())

        train_f1 = f1_score(train_labels, train_preds, average="macro")
        print(f"[Epoch {epoch+1}] Train Loss: {train_loss/len(train_loader):.4f}, F1: {train_f1:.4f}")
```

```python
def take_from_train_logits(model, train_loader, device):
  final_logits = []
  model.eval()
  with torch.no_grad():
    for images in test_loader:
      images = images.to(DEVICE)
      outputs = model(images)
      final_logits.append(outputs)
  return final_logits
```

```python
def eval_model(model, test_loader, device):
  final_logits = []
  model.eval()
  with torch.no_grad():
    for images in test_loader:
      images = images.to(DEVICE)
      outputs = model(images)
      final_logits.append(outputs)

  all_outputs = torch.cat(final_logits, dim=0)
  overall_preds = all_outputs.argmax(dim=1).cpu().numpy()
  return final_logits, overall_preds
```

```python
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

model = EfficientNet(num_classes=8).to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

train_model_without_eval(model, train_full_loader, optimizer, device=DEVICE)
```

```python
logits_by_EN, preds_by_EN = eval_model(model, test_loader, DEVICE)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model_bert = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=8)
```

```python
def clean_text_basic(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    return text



X_train_for_bert = X_train.drop(columns=['id', 'label'])
X_train_for_bert['label'] = X_train['mapped']
X_train_for_bert = X_train_for_bert.drop(columns=['mapped'])
X_train_for_bert['description'] = X_train['description'].map(clean_text_basic)
```

```python
X_test_for_bert = test.drop(columns=['id'])
X_test_for_bert = X_test_for_bert.map(clean_text_basic)
```

```python
def tokenize(batch):
  return tokenizer(batch['description'], truncation=True, padding='max_length', max_length=64)
```

```python
from datasets import Dataset

train_dataset = Dataset.from_pandas(X_train_for_bert)
test_dataset = Dataset.from_pandas(X_test_for_bert)
```

```python
train_dataset_tokenized = train_dataset.map(tokenize, batched=True)
test_dataset_tokenized = test_dataset.map(tokenize, batched=True)
```

```python
args = TrainingArguments(
    output_dir='xz',
    logging_steps=250,
    per_device_train_batch_size=32,
    learning_rate=3e-5,
    logging_dir='none',
    report_to='none',
    fp16=True,
    save_strategy='no',
    num_train_epochs=5,
    weight_decay=0.01
    )
```

```python
from sklearn.metrics import f1_score
```

```python
def compute_metrics(outdess):
  logits, labels = outdess
  preds = logits.argmax(-1)
  return {'f1_macro': f1_score(labels, preds, average='macro')}
```

```python
trainer = Trainer(
    model=model_bert,
    args=args,
    train_dataset=train_dataset_tokenized,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

# Submission engineering

### крутая функция alpha - вверх, важность логитов cv модели - вниз

```python
import torch
import numpy as np
import pandas as pd

def to_tensor(data):
    """Преобразует входные данные в torch.FloatTensor на CPU."""
    if isinstance(data, list):
        # Список тензоров → склеиваем
        if isinstance(data[0], torch.Tensor):
            data = torch.cat(data, dim=0)
        else:
            data = torch.tensor(data)
    elif isinstance(data, pd.DataFrame):
        data = torch.tensor(data.values)
    elif isinstance(data, np.ndarray):
        data = torch.from_numpy(data)
    elif isinstance(data, torch.Tensor):
        data = data
    else:
        raise TypeError(f"Unsupported type: {type(data)}")

    return data.detach().cpu().float()

def weighted_averaging(logits_by_Resnet, logits_by_BERT, logits_by_EN, alpha=0.5, beta=0.3, gamma=0.2):
    logits_by_Resnet = to_tensor(logits_by_Resnet)
    logits_by_BERT   = to_tensor(logits_by_BERT)
    logits_by_EN     = to_tensor(logits_by_EN)

    combined = alpha * logits_by_BERT + beta * logits_by_Resnet + gamma * logits_by_EN
    return combined

```

```python
combined_logits = weighted_averaging(logits_by_RN, logits_by_bert, logits_by_EN)
```

```python
predictions_overall = combined_logits.argmax(1)
predictions_done = pd.Series(predictions_overall).map(func_back)
```
