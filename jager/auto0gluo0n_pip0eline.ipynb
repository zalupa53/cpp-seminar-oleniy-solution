## üìö –ò–º–ø–æ—Ä—Ç—ã

```python
import pandas as pd
import numpy as np
from autogluon_pipeline import AutoGluonPipeline

# –î–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
import matplotlib.pyplot as plt
import seaborn as sns

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-darkgrid')

print("‚úÖ –ò–º–ø–æ—Ä—Ç—ã –≤—ã–ø–æ–ª–Ω–µ–Ω—ã —É—Å–ø–µ—à–Ω–æ!")
```

# üéØ –ü—Ä–∏–º–µ—Ä 1: –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∫–æ–¥ (3 —Å—Ç—Ä–æ–∫–∏!)

–°–∞–º—ã–π –±—ã—Å—Ç—Ä—ã–π —Å–ø–æ—Å–æ–± –Ω–∞—á–∞—Ç—å —Ä–∞–±–æ—Ç—É —Å AutoGluon.

```python
# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–µ–º–æ-–¥–∞–Ω–Ω—ã–µ –∏–∑ AutoGluon
from autogluon.tabular import TabularDataset

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –¥–∞—Ç–∞—Å–µ—Ç –æ –¥–æ—Ö–æ–¥–∞—Ö
train_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv')
test_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv')

# –ë–µ—Ä–µ–º –ø–æ–¥–≤—ã–±–æ—Ä–∫—É –¥–ª—è –±—ã—Å—Ç—Ä–æ–π –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
train_data = train_data.sample(n=1000, random_state=42)
test_data = test_data.sample(n=500, random_state=42)

print(f"üìä –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã!")
print(f"   –¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞: {train_data.shape}")
print(f"   –¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞: {test_data.shape}")
print(f"\nüéØ –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è: class (–¥–æ—Ö–æ–¥ >50K –∏–ª–∏ <=50K)")
```

```python
# –°–º–æ—Ç—Ä–∏–º –Ω–∞ –¥–∞–Ω–Ω—ã–µ
train_data.head()
```

```python
# üöÄ –í–°–Å –ó–ê 3 –°–¢–†–û–ö–ò!

# 1. –°–æ–∑–¥–∞–µ–º pipeline
pipeline = AutoGluonPipeline(label='class', eval_metric='roc_auc')

# 2. –û–±—É—á–∞–µ–º (2 –º–∏–Ω—É—Ç—ã)
pipeline.fit(train_data, time_limit=120, presets='medium_quality')

# 3. –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º
predictions = pipeline.predict(test_data.drop(columns=['class']))

print("\n‚úÖ –ì–æ—Ç–æ–≤–æ! –ú–æ–¥–µ–ª—å –æ–±—É—á–µ–Ω–∞ –∏ –¥–µ–ª–∞–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è.")
```

```python
# –°–º–æ—Ç—Ä–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
print("üîÆ –ü–µ—Ä–≤—ã–µ 10 –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π:")
print(predictions.head(10))
```

---

# üìä –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏

```python
# –ú–µ—Ç—Ä–∏–∫–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
metrics = pipeline.evaluate(test_data)

print("üìè –ú–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:")
print("="*50)
for metric, value in metrics.items():
    print(f"  {metric:25s}: {value:.4f}")
```

```python
# –¢–∞–±–ª–∏—Ü–∞ –ª–∏–¥–µ—Ä–æ–≤ - –∫–∞–∫–∏–µ –º–æ–¥–µ–ª–∏ –æ–±—É—á–µ–Ω—ã
leaderboard = pipeline.leaderboard(test_data)

print("\nüèÜ –¢–∞–±–ª–∏—Ü–∞ –ª–∏–¥–µ—Ä–æ–≤ (—Ç–æ–ø-5 –º–æ–¥–µ–ª–µ–π):")
print("="*50)
display(leaderboard.head())
```

```python
# –í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
importance = pipeline.feature_importance(test_data)

print("\nüéØ –í–∞–∂–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (—Ç–æ–ø-10):")
print("="*50)
display(importance.head(10))
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
plt.figure(figsize=(10, 6))
top_features = importance.head(10)
plt.barh(range(len(top_features)), top_features['importance'])
plt.yticks(range(len(top_features)), top_features.index)
plt.xlabel('–í–∞–∂–Ω–æ—Å—Ç—å')
plt.title('–¢–æ–ø-10 –≤–∞–∂–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

---

# üîÆ –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏

```python
# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∞
probabilities = pipeline.predict_proba(test_data.drop(columns=['class']))

print("üìà –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ (–ø–µ—Ä–≤—ã–µ 10):")
display(probabilities.head(10))
```

```python
# –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
results = test_data.copy()
results['prediction'] = predictions
results['prob_high_income'] = probabilities.iloc[:, 1]
results['correct'] = (results['class'] == results['prediction'])

print(f"‚úÖ –¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π: {results['correct'].mean():.2%}")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø—Ä–∏–º–µ—Ä—ã
print("\nüìä –ü—Ä–∏–º–µ—Ä—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π:")
display(results[['age', 'education', 'occupation', 'class', 'prediction', 'prob_high_income', 'correct']].head(10))
```

```python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(probabilities.iloc[:, 1], bins=50, alpha=0.7, edgecolor='black')
plt.xlabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Å–æ–∫–æ–≥–æ –¥–æ—Ö–æ–¥–∞')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π')
plt.axvline(0.5, color='red', linestyle='--', label='–ü–æ—Ä–æ–≥ 0.5')
plt.legend()

plt.subplot(1, 2, 2)
correct_probs = results[results['correct']]['prob_high_income']
incorrect_probs = results[~results['correct']]['prob_high_income']
plt.hist(correct_probs, bins=30, alpha=0.5, label='–ü—Ä–∞–≤–∏–ª—å–Ω—ã–µ', color='green')
plt.hist(incorrect_probs, bins=30, alpha=0.5, label='–û—à–∏–±–æ—á–Ω—ã–µ', color='red')
plt.xlabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã—Å–æ–∫–æ–≥–æ –¥–æ—Ö–æ–¥–∞')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.title('–£–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏')
plt.legend()

plt.tight_layout()
plt.show()
```

# üéØ –ü—Ä–∏–º–µ—Ä 2: –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏

AutoGluon –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏!

```python
# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏
df_with_missing = pd.DataFrame({
    'age': [25, 30, None, 45, 50, 35, None, 28, 42, 38],
    'city': ['Moscow', 'SPB', 'Moscow', None, 'Kazan', 'Moscow', 'SPB', 'Kazan', 'Moscow', 'SPB'],
    'income': [50000, 60000, 55000, 70000, None, 65000, 58000, 52000, 72000, 61000],
    'has_car': [True, False, True, True, None, False, True, False, True, True],
    'education': ['Bachelor', 'Master', 'Bachelor', 'PhD', 'Master', None, 'Bachelor', 'Master', 'PhD', 'Bachelor'],
    'credit_score': [700, 750, None, 800, 720, 680, None, 710, 790, 740],
    'approved': [1, 1, 1, 1, 0, 1, 0, 1, 1, 1]  # –û–¥–æ–±—Ä–µ–Ω –ª–∏ –∫—Ä–µ–¥–∏—Ç
})

print("üìä –î–∞–Ω–Ω—ã–µ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏:")
display(df_with_missing)

print(f"\nüîç –ü—Ä–æ–ø—É—Å–∫–∏ –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º:")
print(df_with_missing.isnull().sum())
```

```python
# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –ë–ï–ó –ö–ê–ö–û–ô-–õ–ò–ë–û –ü–†–ï–î–û–ë–†–ê–ë–û–¢–ö–ò!
pipeline_missing = AutoGluonPipeline(label='approved')
pipeline_missing.fit(df_with_missing, time_limit=60, presets='medium_quality')

print("\n‚úÖ –ú–æ–¥–µ–ª—å –æ–±—É—á–µ–Ω–∞ —É—Å–ø–µ—à–Ω–æ!")
print("\nAutoGluon –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏:")
print("  ‚úì –û–±—Ä–∞–±–æ—Ç–∞–ª –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è")
print("  ‚úì –ó–∞–∫–æ–¥–∏—Ä–æ–≤–∞–ª –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏")
print("  ‚úì –û–ø—Ä–µ–¥–µ–ª–∏–ª –±—É–ª–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏")
```

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö (—Ç–æ–∂–µ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏!)
new_data = pd.DataFrame({
    'age': [35, None, 42],
    'city': ['Moscow', 'SPB', None],
    'income': [65000, 58000, 72000],
    'has_car': [True, None, False],
    'education': ['Bachelor', 'Master', None],
    'credit_score': [690, 740, None]
})

print("üîÆ –ù–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è (—Ç–æ–∂–µ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏):")
display(new_data)

predictions_missing = pipeline_missing.predict(new_data)
proba_missing = pipeline_missing.predict_proba(new_data)

print("\nüìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
results_missing = new_data.copy()
results_missing['approved_prob'] = proba_missing.iloc[:, 1]
results_missing['prediction'] = predictions_missing
display(results_missing[['approved_prob', 'prediction']])
```

# üèãÔ∏è –ü—Ä–∏–º–µ—Ä 3: –†–µ–≥—Ä–µ—Å—Å–∏—è - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Ü–µ–Ω

AutoGluon —Ç–∞–∫–∂–µ –æ—Ç–ª–∏—á–Ω–æ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å –∑–∞–¥–∞—á–∞–º–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏!

```python
# –°–æ–∑–¥–∞–µ–º —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ –æ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=500, n_features=10, noise=10, random_state=42)

df_regression = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(10)])
df_regression['price'] = y * 1000 + 500000  # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Ü–µ–Ω—ã

# –†–∞–∑–¥–µ–ª—è–µ–º –Ω–∞ train/test
train_reg = df_regression[:400]
test_reg = df_regression[400:]

print(f"üìä –î–∞–Ω–Ω—ã–µ –æ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏ —Å–æ–∑–¥–∞–Ω—ã")
print(f"   –¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞: {train_reg.shape}")
print(f"   –¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞: {test_reg.shape}")
print(f"\nüí∞ –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ü–µ–Ω:")
print(train_reg['price'].describe())
```

```python
# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –¥–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
pipeline_regression = AutoGluonPipeline(
    label='price',
    eval_metric='r2'  # R¬≤ Score –¥–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
)

pipeline_regression.fit(
    train_reg,
    time_limit=120,
    presets='medium_quality'
)

print("‚úÖ –ú–æ–¥–µ–ª—å –¥–ª—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –æ–±—É—á–µ–Ω–∞!")
```

```python
# –û—Ü–µ–Ω–∏–≤–∞–µ–º –∫–∞—á–µ—Å—Ç–≤–æ
metrics_reg = pipeline_regression.evaluate(test_reg)

print("üìè –ú–µ—Ç—Ä–∏–∫–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:")
print("="*50)
print(f"  R¬≤ Score:  {metrics_reg['r2']:.4f}")
print(f"  MAE:       ${-metrics_reg['mean_absolute_error']:,.2f}")
print(f"  RMSE:      ${-metrics_reg['root_mean_squared_error']:,.2f}")
```

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º —Ü–µ–Ω—ã
predicted_prices = pipeline_regression.predict(test_reg.drop(columns=['price']))
actual_prices = test_reg['price'].values

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(actual_prices, predicted_prices, alpha=0.6)
plt.plot([actual_prices.min(), actual_prices.max()], 
         [actual_prices.min(), actual_prices.max()], 
         'r--', lw=2, label='–ò–¥–µ–∞–ª—å–Ω–æ–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ')
plt.xlabel('–†–µ–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞')
plt.ylabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è —Ü–µ–Ω–∞')
plt.title('–ö–∞—á–µ—Å—Ç–≤–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π')
plt.legend()

plt.subplot(1, 2, 2)
errors = predicted_prices - actual_prices
plt.hist(errors, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='–ù—É–ª–µ–≤–∞—è –æ—à–∏–±–∫–∞')
plt.xlabel('–û—à–∏–±–∫–∞ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—à–∏–±–æ–∫')
plt.legend()

plt.tight_layout()
plt.show()

print(f"\nüìä –°—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞: ${abs(errors).mean():,.2f}")
```

# üé® –ü—Ä–∏–º–µ—Ä 4: –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞–∑–Ω—ã—Ö –ø—Ä–µ—Å–µ—Ç–æ–≤

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –≤–ª–∏—è–µ—Ç –≤—ã–±–æ—Ä –ø—Ä–µ—Å–µ—Ç–∞ –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –∏ –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è.

```python
import time

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –Ω–µ–±–æ–ª—å—à—É—é –≤—ã–±–æ—Ä–∫—É –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
train_small = train_data.sample(n=500, random_state=42)
test_small = test_data.sample(n=200, random_state=42)

presets_to_test = ['medium_quality', 'good_quality']
results_comparison = []

for preset in presets_to_test:
    print(f"\n{'='*60}")
    print(f"–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–µ—Å–µ—Ç–∞: {preset}")
    print('='*60)
    
    start_time = time.time()
    
    # –û–±—É—á–∞–µ–º
    pipeline_preset = AutoGluonPipeline(
        label='class',
        path=f'models_{preset}'
    )
    
    pipeline_preset.fit(
        train_small,
        time_limit=180,  # 3 –º–∏–Ω—É—Ç—ã –Ω–∞ –∫–∞–∂–¥—ã–π
        presets=preset
    )
    
    train_time = time.time() - start_time
    
    # –û—Ü–µ–Ω–∏–≤–∞–µ–º
    metrics_preset = pipeline_preset.evaluate(test_small, silent=True)
    
    results_comparison.append({
        'preset': preset,
        'accuracy': metrics_preset['accuracy'],
        'roc_auc': metrics_preset['roc_auc'],
        'train_time': train_time
    })
    
    print(f"‚úÖ Accuracy: {metrics_preset['accuracy']:.4f}")
    print(f"‚úÖ ROC-AUC: {metrics_preset['roc_auc']:.4f}")
    print(f"‚è±Ô∏è  –í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è: {train_time:.1f} —Å–µ–∫")
```

```python
# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
comparison_df = pd.DataFrame(results_comparison)

print("\n" + "="*60)
print("–ò–¢–û–ì–û–í–û–ï –°–†–ê–í–ù–ï–ù–ò–ï –ü–†–ï–°–ï–¢–û–í")
print("="*60)
display(comparison_df)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Accuracy
axes[0].bar(comparison_df['preset'], comparison_df['accuracy'], color='skyblue')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('–¢–æ—á–Ω–æ—Å—Ç—å')
axes[0].set_ylim([0.7, 0.9])

# ROC-AUC
axes[1].bar(comparison_df['preset'], comparison_df['roc_auc'], color='lightgreen')
axes[1].set_ylabel('ROC-AUC')
axes[1].set_title('ROC-AUC')
axes[1].set_ylim([0.7, 0.9])

# –í—Ä–µ–º—è
axes[2].bar(comparison_df['preset'], comparison_df['train_time'], color='coral')
axes[2].set_ylabel('–°–µ–∫—É–Ω–¥—ã')
axes[2].set_title('–í—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è')

plt.tight_layout()
plt.show()
```
